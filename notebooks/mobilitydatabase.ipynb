```python
from dotenv import load_dotenv
import os
import requests

load_dotenv()
api_key = os.getenv("mobility_database_token")
```

```python
limit = 1000
offset = 0
# options are vp, tu, and sa
entity_types = ["vp"]
data = True
feeds = []

while data:
    url = f"https://api.mobilitydatabase.org/v1/gtfs_rt_feeds?limit={limit}&offset={offset}&entity_types={','.join(entity_types)}"
    headers = {"accept": "application/json", "Authorization": f"Bearer {api_key}"}
    response = requests.get(url=url, headers=headers)
    if response.status_code != 200:
        data = False
    if len(response.json()) == 0:
        data = False
    print(len(response.json()))
    data_entries = response.json()
    feeds.extend(data_entries)
    offset += limit
```

```python
feeds
```

```python
import pandas as pd
```

```python
df = pd.json_normalize(feeds)
```

```python
df
```

```python
authenticated_feeds = df[df["source_info.authentication_type"] > 0]
```

```python
authenticated_feeds.to_csv("authenticated_feeds.csv")
```
